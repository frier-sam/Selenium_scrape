# 1 Link scraping

```python
from villa_scrape import link_scrape
url = "https://www.kompas-villas.com/Croatia/villas-croatia/EN?region=&arrival-date=&departure-date=&person=&name=&TipObjektaId="
sleeptime = 7 # change it based on the speed your internet connection (page need to load totally before scraping)
pages = 2 #no.of pages to scrape (you may enter 100 or 200 and if the script can't find the page it will not throw a error)
lscrape = link_scrape.scrape(url,pages,sleeptime)
```

Saving the links in csv for future use

```python
import csv
with open('links.csv', 'w') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerow(lscrape.links)
```

# 2. Item Scrape

```python
from villa_scrape import item_scrape
sleeptime = 7
iscrape = item_scrape.scrape(lscrape.links,sleeptime)
```
